```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2> Membaca dataset yang telah di download dengan format csv</h2>


```python
df = pd.read_csv("Indonesia_coronavirus_daily_data.csv")
```

<h2>df untuk menampilkan isi yang ada di dalam dataset</h2>

```python
df
```

<h2>Menampilkan type dari atribut yang ada di dalam dataset</h2>

```python
df.dtypes
```

<h2>Menampilkan jumlah isi dari atribut Province</h2>

```python
df['Province'].value_counts()
```

<h2>Membuat tabel untuk atribut Province dan membuat atribut jumlah untuk atribut di dalamnya</h2>

```python
labels = df['Province'].value_counts().rename_axis('Province').reset_index(name='Jumlah')
labels
```

<h2>Membuat grafik pie</h2>

```python
plt.figure(figsize=(20,20), dpi=400)
plt.pie(df['Province'].value_counts(), labels=labels['Province'], autopct='%.1f%%', startangle=90)
plt.show()
```

<h2>Grafik Boxplot </h2>

```python
sns.boxplot(x= df['Cumulative_Active_Case'], y=df['Province'])
plt.show()
```

<h2>Grafik Scatter Plot</h2>

```python
sns.scatterplot(data = df, x='Cumulative_Active_Case', y = 'Province')
plt.title("Persebaran data ukuran Attack dengan Attribute yang beda")
plt.show()
```
